```python
#import dataframe with countrycode
df_country = pd.read_excel('Country-code.xlsx')
```

```python
df_country.head(50)
```

```python
# merge countrycode to country in original data
df_new = pd.merge(df,df_country,on = 'Country Code',how = 'left')
df_new.columns
```

```python
df_new.groupby(['Country','Has Online delivery']).size().reset_index()
```

```python
#separate numerical and categorical data from data
df_numeric_cols = df_new.columns[df_new[df_new.columns].dtypes != 'O']
df_cat_cols = df_new.columns[[df_new[df_new.columns].dtypes == 'O']]
df_numeric = df_new[df_numeric_cols]
df_cat = df_new[df_cat_cols]
```

```python
# check countries represented 
#df_new.groupby('Country')['Country'].count()
country_count = df_new.value_counts('Country')
country_count
```

```python
plt.figure(figsize=(15,6))
sns.countplot(data = df_new, y = 'Country',)
```

```python
plt.figure(figsize=(6,6))
plt.pie(data= df_cat_cols, x = country_count.values[:3], labels=country_count.index[:3],autopct= '%.2f%%')
```

Observation;
India, US and Uk ae the most populated country in our dataset

```python
agg_rating_mean = df_new.groupby(['Country'])['Aggregate rating'].mean().values
agg_rating_index = df_new.groupby(['Country'])['Aggregate rating'].mean().index
```

```python
plt.figure(figsize=(12,12))
plt.pie(data= df_cat_cols, x = agg_rating_mean, labels=agg_rating_index,autopct= '%.2f%%')
```
